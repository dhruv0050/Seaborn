```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset("tips")
df.head()
```

Bar Plot

```python
sns.barplot(x="day", y="total_bill", data=df, color="blue", estimator=np.median)
plt.title("Bar Plot of Total Bill by Day")
```

Count Plot

```python
sns.countplot(x='sex', data=df, palette='Set2')
```

Box Plot

```python
sns.boxplot(x='day',y='total_bill', data=df, palette='Set1',hue='sex')
plt.legend(loc=0)
```

Violin Plot

```python
sns.violinplot(x='day',y='total_bill', data=df, palette='Set2',hue='time', split=True)
plt.legend(loc=0)
```

Strip Plot

```python
sns.stripplot(x='day', y='total_bill', data=df, jitter=True, hue='sex',palette='Set1')
plt.legend(loc=0)
```

```python
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=df, jitter=True, hue='time',palette='afmhot', dodge=True)
plt.legend(loc=1)
```

Swarm Plot

```python
sns.swarmplot(x='day',y='total_bill',data=df,palette='cividis')
```

```python
sns.swarmplot(x='day',y='total_bill',data=df,palette='magma',hue='time')
plt.legend(loc=0)
```
