```python
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
df.head(4)
```

```python
sns.jointplot(x = 'ins_losses', y ='speeding', data = df, kind = 'reg', color = 'red')
```

```python
sns.jointplot(x='speeding',y='alcohol', data=df, kind='hex', color='green')
```

```python
sns.jointplot(x='no_previous',y='ins_premium', data=df, kind='kde')
```
