```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
df = sns.load_dataset('car_crashes')
df.head(4)
```

Heat Maps -- 
Way-A : Creating a correlation

```python
mx = df.drop(columns=['abbrev']).corr()
sns.heatmap(mx,annot=True,cmap='coolwarm')
```

Heat Map -- Way - B : creating a pivot table

```python
f_df = sns.load_dataset('flights')
f_df.head(4)
```

```python
f_mx = f_df.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(f_mx,linecolor='white',linewidths=1,cmap='Blues')
```

Cluster Map

```python
sns.clustermap(f_mx,linecolor='white',linewidths=1,cmap='Blues',standard_scale=1)
```

Pair Grid

```python
iris = sns.load_dataset('iris')
iris.head(4)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map(plt.scatter)
```

```python
i = sns.PairGrid(iris,hue='species')
i.map_diag(sns.kdeplot)
i.map_offdiag(plt.scatter)
```

```python
j = sns.PairGrid(iris,hue='species')
j.map_diag(sns.violinplot)
j.map_upper(sns.kdeplot)
j.map_lower(plt.scatter)
```
