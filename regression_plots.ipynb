```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('car_crashes')
df.head(4)
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head(4)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips_df,hue='sex',markers=['o','s'],palette='afmhot')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips_df,col='sex',row='time')
```
